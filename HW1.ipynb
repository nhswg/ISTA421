## Q6

```python
weight = np.random.normal(81, 0.26, 100)
height = np.random.normal(1.76, 0.22, 100)
age = np.random.normal(25, 5, 100)

gender = np.random.choice(['A', 'B', 'C'], 100)

print('weight:', weight)
print('height:', height)
print('age:', age)
print('gender:', gender)
```

## Q7
source again : https://www.statology.org/add-numpy-array-to-pandas-dataframe/

```python
Q7 = pd.DataFrame({
    'weight': weight,
    'height': height,
    'age': age,
})
BMI = np.nan
Q7['BMI'] = BMI

print(Q7)
```

## Q8

```python
Q8 = Q7.copy()
AgeGroup = np.where((Q8['age'] < 20) | (Q8['age'] > 30), True, False)
Q8['AgeGroup'] = AgeGroup
print(Q8)
```

## Q9

```python
Q9 = Q8.copy()
Q9['BMI'] = Q9['weight'] / (Q9['height'] ** 2)
print(Q9)
```

## Q10
source : https://pandas.pydata.org/docs/user_guide/indexing.html

```python
def function_Q10(weight, height):
    BMI = weight / (height ** 2)
    return BMI

test_function_Q10 = function_Q10(Q9['weight'][99], Q9['height'][99])
print(test_function_Q10)
```

## Q11

```python
def function_Q11(df, target_gender):
    filtered = df[df['gender'] == target_gender] # Filter for target gender
    return filtered['BMI'].mean() # Return mean BMI for gender

Q9_with_gender = Q9.copy()
Q9_with_gender['gender'] = gender
test_function_Q11 = function_Q11(Q9_with_gender, 'A')
print(test_function_Q11)
```
